P(SS<문장>SE) = P(시작단어|SS) * P(|)* P(|)* P(|) *P(SE|종료단어)

```python
import numpy as np
def s_sc_f(x): #바이그램 언어 모형의 확률 연산 함수
    p=0.0
    for i in range(len(x)-1):
        c=x[i]
        w=x[i+1]
        p += np.log(cpd[c].prob(w)-np.finfo(float).eps) #확률이 더해짐
    return np.exp(p)
t_data=['the','movie','.']
t_data2=['movie','.','the']
s_sc_f(t_data), s_sc_f(t_data2)
```

```python
data_l = []
for sentence in data:
    bg = ngrams(sentence, 2, pad_left=True, pad_right = True, left_pad_symbol='SS', right_pad_symbol='SE')
    data_l += [tk for tk in bg]
#바이그램으로 토큰화가 됐음
#문맥을 이해하기 위해 단어끼리의 연관성을 생각하는 것
```

```python
import random
def 정리_생성(c='SS', s_n=10): #시작 단어와 랜덤 시드를 직접 정할 수 있다.
    random.seed(s_n)
    all_data=[]
    while True:
        if c not in cpd: #시작 단어가 존재하지 않는다면 그냥 break
            break
        w=cpd[c].generate()
        
        #right pad로 설정한 문자열 
        if w == 'SE':
            break
        #이후 단어가 어떤 단어인지에 따라 대문자 
        elif w in ['i','ii','iii']: #시작 단어라는 키워드.
            w2=w.upper()
        elif w in ['mr', 'luc', 'i', 'robin', 'williams', 'cindy', 'crawford']: #아무튼 시작되는 키워드..
            w2 = w.title() #맨 앞글자만 대문자로
        else:
            w2 = w
        
        #리스트에 넣기 전 어떤 문자인지 확인
        if c == 'SS':
            all_data.append(w2.title())
        
        #앞글자가 아래의 문자로 시작한다면 그냥 넣음
        elif c in ["'", "\"", '(']:
            all_data.append(w2)
        #뒷글자가 아래의 문자로 시작한다면 그냥 넣음
        elif w in ["'", "\'", ",", ".", ")", ":", ";", "?"]:
            all_data.append(w2)
        #그밖의 것들은 공백과 함께 넣어서 문장처럼 보이도록 한다.
        else:
            all_data.append(' '+w2)
        #리스트에 append한 후 현재의 뒷글자를 앞글자로 바꾸어주고 while문 다시 시작
        c=w
    #위의 while문이 끝났을 때 리스트가 비어있지 않다면 하나의 문자열로 연결해서 return
    if all_data:
        return ''.join(all_data)
    #위의 조건문을 통과했다는 것은 비어있는 리스트. 아무 문자열도 없기 때문에 아래의 문자열을 return
    return '문장을 생성할 수 없습니다.'
```
