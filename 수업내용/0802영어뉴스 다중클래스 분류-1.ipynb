```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
w_data = fetch_20newsgroups()
w_data.keys()
```

```python
X_data = w_data.data
y_data = w_data.target
```

```python
import numpy as np 
np.unique(y_data) #다중 클래스 분류
```

```python
all_data = pd.DataFrame()
all_data['X'] = X_data
all_data['y'] = y_data
all_data
```

```python
n_l = []
for i in X_data:
    n = len(i)
    n_l.append(n)
max(n_l), min(n_l)
```

```python
tk = Tokenizer(num_words=10000) #차원을 정함.
tk.fit_on_texts(X_data)
t_x_data = tk.texts_to_matrix(X_data)
t_x_data.shape
```

```python
t_x_data
```

```python
t1_x_data = tk.texts_to_matrix(X_data, mode='count')
t1_x_data
```

```python
t2_x_data = tk.texts_to_matrix(X_data, mode='tfidf')
t2_x_data
```

```python
X_data1 = t_x_data
X_data2 = t1_x_data
X_data3 = t2_x_data
```

```python
Y_data = to_categorical(y_data)
```

```python
len(X_data1),len(X_data2),len(X_data3),len(Y_data)
```

```python
from sklearn.model_selection import train_test_split
t_x1, tt_x1, t_y1, tt_y1 = train_test_split(X_data1, Y_data, random_state=1) #단어 유무 특징 갖는 data
t_x2, tt_x2, t_y2, tt_y2 = train_test_split(X_data2, Y_data, random_state=1) #단어 빈도수 data
t_x3, tt_x3, t_y3, tt_y3 = train_test_split(X_data3, Y_data, random_state=1) #tdidf 단어 가중치 data
```

모델 생성

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
m = Sequential()
m.add(Dense(256,input_shape=(10000,), activation='relu'))
m.add(Dropout(0.3))
m.add(Dense(128, activation='relu'))
m.add(Dropout(0.3))
m.add(Dense(20, activation='softmax'))
m.compile(optimizer='adam',loss='categorical_crossentropy', metrics = 'accuracy')
hy = m.fit(t_x1, t_y1, epochs=10, validation_data=[tt_x1, tt_y1])
```
